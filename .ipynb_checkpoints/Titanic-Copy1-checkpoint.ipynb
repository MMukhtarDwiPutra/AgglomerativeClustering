1. Import dulu librarynya
2. Pretty display?
3. Set random seed (biar ga berubah mungkin)
4. Load dataset
5. Liat datanya duls

```python
#1. Import library
import numpy as np
import pandas as pd
from IPython.display import display #Untuk visualisasi

#2. Pretty display for notebooks
%matplotlib inline

#3. Set random seed
import random
random.seed(42)

#4. Load dataset
input_file = 'titanic_data.csv'
full_data = pd.read_csv(input_file)

#5. Lihat duls
display(full_data.head())
```

1. Kumpulin feature untuk accuracy? dalam variable baru
2. Remove feature untuk accuracy dari dalam dataset

```python
outcomes = full_data['Survived']
features_raw = full_data.drop('Survived', axis = 1)

display(features_raw.head())
```

Preprocessing
1. Di one hot encode
2. Fill blank with zero

```python
features = pd.get_dummies(features_raw)
features = features.fillna(0.0)

display(features.head())
```

Train model
1. import train_test_split dari sklearn.model_selection
2. Split ke training dan testing
3. import tree classifier dari sklearn.tree
4. Bikin model DecisionTreeClassifier
5. di fit model train

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, outcomes, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

Testing modelnya
1. bikin predict dari train dan testing
2. Import accuracy_score dari sklearn.metrics
3. Bikin accuracy_score train dan test

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print('The training accuracy is ',train_accuracy)
print('The test accuracy is ',test_accuracy)
```

Cara improve model
1. train model decision tree pakai : hyperparam(max_depth/min_samples_leaf/min_sample_split)
2. fit modelnya
3. predict model (train, test)
4. tinggal accuracy deh (train,test)

```python
model = DecisionTreeClassifier(max_depth = 6, min_samples_leaf = 6, min_samples_split = 8)
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print("Accuracy train ",train_accuracy)
print("Accuracy test ",test_accuracy)
```

```python
import pandas as pd 

#dataframe
df = pd.DataFrame({
    'x':[12,20,28,18,29,33,24,45,45,52,51,52,55,53,55,61,64,69,72],
    'y':[39,36,30,52,54,46,55,59,63,70,66,63,58,23,14,8,19,7,24]
})

df.head()
```

```python
#agglomorative clustering
from sklearn.cluster import AgglomerativeClustering

#Buat model dengan agglomorative clustering
cluster = AgglomerativeClustering(n_clusters = 3)
cluster.fit(df)
cluster.labels_
```

```python
import matplotlib.pyplot as plt

#Plot
plt.scatter(df['x'],df['y'],c=cluster.labels_,cmap='rainbow')
plt.title('Agglomerative Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Dengan dataset titanic (Implementasi dengan dataset)

```python
df = pd.read_csv('titanic_data.csv')

import matplotlib.pyplot as plt

#agglomorative clustering
from sklearn.cluster import AgglomerativeClustering

fitur = df[['Age','Fare']]

fitur = fitur.dropna()

#Buat model dengan agglomorative clustering
cluster = AgglomerativeClustering(n_clusters = 3)
cluster.fit(fitur)
cluster.labels_

#Plot
plt.scatter(fitur['Age'],fitur['Fare'],c=cluster.labels_,cmap='rainbow')
plt.title('Titanic Data')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
